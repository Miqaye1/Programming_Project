```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/raw/Topic2_credit_assignment_dataset.csv")
list(df)
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
df.head()
```

```python
df_labeled   = df[df['SeriousDlqin2yrs'].notna()].copy()
df_unlabeled = df[df['SeriousDlqin2yrs'].isna()].copy()

print(f"Labeled rows   : {len(df_labeled):,}")
print(f"Unlabeled rows : {len(df_unlabeled):,}")
```

```python
df_labeled['DebtRatio'].describe()
```

```python
df_labeled['DebtRatio'].quantile([0.95, 0.975])
```

```python
for q in [0.95, 0.975]:
    threshold = df_labeled['DebtRatio'].quantile(q)
    subset = df_labeled[df_labeled['DebtRatio'] > threshold]
    total_with_income = subset['MonthlyIncome'].notna().sum()
    binary_income = subset[subset['MonthlyIncome'].isin([0, 1])]['MonthlyIncome'].count()
    print(f"q={q} | threshold={threshold:.1f} | records={len(subset):,} | binary_income={binary_income}/{total_with_income}")
```

```python
for q, threshold in [(0.95, df_labeled['DebtRatio'].quantile(0.95)), (0.975, df_labeled['DebtRatio'].quantile(0.975))]:
    count_1 = df_labeled[(df_labeled['DebtRatio'] > threshold) & (df_labeled['MonthlyIncome'] == 1) & (df_labeled['SeriousDlqin2yrs'] == 1)].shape[0]
    count_0 = df_labeled[(df_labeled['DebtRatio'] > threshold) & (df_labeled['MonthlyIncome'] == 0) & (df_labeled['SeriousDlqin2yrs'] == 0)].shape[0]
    print(f"q={q} | Income=1 & Default=1: {count_1} | Income=0 & Default=0: {count_0}")
```
